## Data Visualization

```python
GP_by = df.groupby(by='Continent').sum()
GP_by
```

```python
a= sns.barplot(x= GP_by["Rank"].index , y='Rank',data=GP_by)
plt.xticks(rotation=45)
a.bar_label(a.containers[0], fontsize=8);
print()
```

```python
# TOP 5 COUNTRIES WITH HIGHEST POPULATION
df_sorted = df.sort_values(by='Rank', ascending=True).head()
df_sorted
```

```python
df_sort = df_sorted[["Country/Territory", "1970 Population","1980 Population","1990 Population","1980 Population","2000 Population","2010 Population","2015 Population","2020 Population","2022 Population"]]
df_sort
```

```python
# POPULATION OF INDIA OVER THE YEAR
x= df_sort.iloc[1,1:]

a= sns.barplot(x= x.index , y=x.values)
plt.title("Population of India")
plt.ylabel("Billions")
plt.xticks(rotation=45)
a.bar_label(a.containers[0], fontsize=6);
print()
```

```python
import matplotlib.pyplot as plt

# Create some data
x = df_sort.iloc[0,1:]
y = df_sort.iloc[1,1:]


# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # Create a figure with 1 row and 2 columns

# Plot data on the first subplot
axes[0].plot(x, x.index,  color='blue', marker='o')
axes[0].set_title('China Population')
axes[0].set_xticklabels(x, rotation=45)

# Plot data on the second subplot
axes[1].plot(y, x.index , color='green', linestyle='--', marker='s')
axes[1].set_title('India Population')
axes[1].set_xticklabels(y, rotation=45)

# Add a title to the entire figure
plt.suptitle('Population of China & India over the decades')

# Adjust layout to prevent overlap
plt.tight_layout()
```

```python
plt.plot(df_sort.iloc[0,1:], color = "red", marker='o', label="China")
plt.plot(df_sort.iloc[1,1:], color = "green" , marker='o' , label= "India")
plt.xticks(rotation = 45)
plt.ylabel("Billions")
plt.title("Population over the decades")
plt.legend()
print()
```
